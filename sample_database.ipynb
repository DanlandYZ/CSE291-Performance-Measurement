```python
import psycopg2
from psycopg2 import sql
import time
import random
from collections import defaultdict
import concurrent.futures
import json
```

```python
host="ec2-52-42-137-44.us-west-2.compute.amazonaws.com"  # Replace with your EC2 instance's public DNS
port = "5432"
dbname = 'dvdrental'
user = "postgres"
password = "postgres"
file_name = 't2.xlarge.data.json'
```

```python
# Define your SQL query
sql_query = """
select
    category.name,
    film.title,
    sum(payment.amount) as "gross revenue"
from
    film
left join film_category on
    film_category.film_id = film.film_id
left join category on
    film_category.category_id = category.category_id
left join inventory on
    inventory.film_id = film.film_id
left join rental on
    rental.inventory_id = inventory.inventory_id
left join payment
    on payment.rental_id = rental.rental_id
where
    payment.amount is not null
group by
    category.name,
    film.title
order by
    sum(payment.amount) desc;
"""

# Connect to the PostgreSQL database
try:
    start_time = time.time()
    conn = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    cursor = conn.cursor()
    
    # Execute the SQL query
    cursor.execute(sql_query)
    
    # Fetch all the results
    results = cursor.fetchall()
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Time taken to execute query : {elapsed_time:.6f} seconds")
    
    print()
    # Print the results
    for row in results:
        print(row)
    
    # Close the cursor and connection
    cursor.close()
    conn.close()
    
except Exception as e:
    print(f"Error: {e}")

```
